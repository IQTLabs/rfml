This notebook demonstrates how to automatically apply a Label to all signals found within a

```python
import sys
sys.path.append("../modules")
import glob

from pathlib import Path
from tqdm import tqdm

import annotation_utils
```

```python
data_directory = "/flash/gamutrf/environment-short/"
file_pattern = "samples_*.sigmf-data"
label = "background"
```

```python
for f in tqdm(glob.glob(str(Path(data_directory,file_pattern)))):
    annotation_utils.annotate(f, label=label, avg_window_len=256, avg_duration=3, debug=True)
```
