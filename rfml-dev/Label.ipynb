```python
import glob

from pathlib import Path
from tqdm import tqdm

import annotation_utils
```

```python
data_globs = {
    "anom_wifi": [
        "data/gamutrf/gamutrf-sd-gr-ieee-wifi/test_offline/anom*.zst",
    ],
    "wifi": [
        "data/gamutrf/gamutrf-sd-gr-ieee-wifi/test_offline/wifi*.zst",
    ],
}
```

```python
for label in data_globs:
    for data_glob in data_globs[label]:
        for f in tqdm(glob.glob(str(Path(data_glob)))):
            annotation_utils.annotate(f, label=label, avg_window_len=256, avg_duration=3, debug=True)
```
