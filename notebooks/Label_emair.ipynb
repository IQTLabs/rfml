```python
import glob
from tqdm import tqdm

import data as data_class
```

```python
def annotate_single_capture(dataset_glob, label):
    for filename in tqdm(glob.glob(dataset_glob+"*.sigmf-meta", recursive=True)):

        data_obj = data_class.Data(filename)
        iq_samples = data_obj.get_samples()
    
        data_obj.sigmf_obj._metadata[data_obj.sigmf_obj.ANNOTATION_KEY] = []
        
        start = 0
        length = 400 #len(iq_samples)
        
        metadata = {
            "core:label": label,
            "core:freq_lower_edge": data_obj.metadata["captures"][0]["core:frequency"]
            - data_obj.metadata["global"]["core:sample_rate"] / 2,
            "core:freq_upper_edge": data_obj.metadata["captures"][0]["core:frequency"]
            + data_obj.metadata["global"]["core:sample_rate"] / 2,
        }
        data_obj.sigmf_obj.add_annotation(start, length=length, metadata=metadata)
    
        data_obj.sigmf_obj.tofile(data_obj.sigmf_meta_filename, skip_validate=True)
```

```python
emair_data = {
    # "anom_wifi": "data/gamutrf/emair_wifi_05032023/20msps/mod/**/",
    # "wifi": "data/gamutrf/emair_wifi_05032023/20msps/normal/**/",
    # "other": "data/gamutrf/emair_wifi_05032023/20msps/other/**/",
    "anom_wifi": "data/gamutrf/wifi-data-03082024/20msps/mod/**/",
    "wifi": "data/gamutrf/wifi-data-03082024/20msps/normal/**/",
}
```

```python
for label in tqdm(emair_data):
    annotate_single_capture(emair_data[label], label)
    
```
